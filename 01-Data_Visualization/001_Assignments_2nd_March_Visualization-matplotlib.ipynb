#### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data to plot a scatter plot.

`import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
Note: Also add title, xlabel, and ylabel to the plot.`

A scatter plot is a graphical representation used to explain the relationship between two continuous variables in a dataset.

It uses a Cartesian coordinate system, where each data point is symbolized by a marker on a two-dimensional plane. 

It represents the independent variable on the x-axis, and the dependent variable on the y-axis. 

By plotting multiple data points, it enables analysts and reserachers to observes the dispersion and clustering of data, identifying potential relationships such as positive or negative correlations, clusters, or outliers. 

```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(3)
x = 3 + np.random.normal(0,2,50)
y = 3 + np.random.normal(0,2,len(x))
```

```python
plt.figure(figsize = (5,5))
plt.scatter(x,y, c = '#3648E1', alpha = 0.8)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("ScatterPlot")
plt.show()
```

#### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function. Use the following data:
`
import numpy as np
For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150]) `

#### Answer:
The `matplotlib.pyplot.subplots()` method provides a way to plot multiple plots on a single figure. Given the number of `rows` and `columns`, it returns a tuple (`fig`, `ax`), giving a single figure `fig` with an array of axes `ax`.

Here is an explanation of the tuple returned by the function:

`fig`: The `matplotlib.pyplot.figure` object to be used as a container for all the subplots. 
`ax`: A single object of the `axes.Axes` object if there is only one plot, or an array of `axes.Axes` objects if there are multiple plots, as specifed by `nrows` and `ncols` arguments. 

```python
# For line plot 1
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

# For line plot 2
x2 = np.array([0, 1, 2, 3, 4, 5]) 
y2 = np.array([50, 20, 40, 20, 60, 70])

# For line plot 3
x3 = np.array([0, 1, 2, 3, 4, 5]) 
y3 = np.array([10, 20, 30, 40, 50, 60])

# For line plot 4
x4 = np.array([0, 1, 2, 3, 4, 5]) 
y4 = np.array([200, 350, 250, 550, 450, 150])

fig, ax = plt.subplots(2,2)
fig.subplots_adjust(hspace = 0.8, top = 0.85, wspace = 0.6)

fig.suptitle("Line Plots", fontsize = 15, ha = 'center')

ax[0,0].plot(x1,y1, color = 'red') # row 0, col = 0
ax[0,0].title.set_text("For line plot 1")
ax[0,0].set_xlabel("X1")
ax[0,0].set_ylabel("Y1")

ax[0,1].plot(x2,y2, color = 'blue') # row 0, col = 1
ax[0,1].title.set_text("For line plot 2")
ax[0,1].set_xlabel("X2")
ax[0,1].set_ylabel("Y2")

ax[1,0].plot(x3,y3, color = 'black') # row 1, col = 0
ax[1,0].title.set_text("For line plot 3")
ax[1,0].set_xlabel("X3")
ax[1,0].set_ylabel("Y3")

ax[1,1].plot(x4,y4, color = 'green') # row 1, col = 1
ax[1,1].title.set_text("For line plot 4")
ax[1,1].set_xlabel("X4")
ax[1,1].set_ylabel("Y4")
plt.show()
```

#### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
`import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000]) `

#### Answer:

A barplot or bar chart or bar graph is a chart or graph that represents categorical data with rectangular bars of equal width with heights or lengths proportional to values that they represent. The bars can be plotted vertically or horizontally. A vertical bar chart is sometimes called a `column chart`
 

Uses of barplot:

1. It provides a visual presentation of categorical variables, discrete variables, or continuous variables grouped in class intervals. 
2. It shows the relationship between a numeric and a categorical variable. 
3. It is a way of summarizing a set a categorical data (continuous data can be made categorical by auto-binning). 
4. Bar charts  arranged from highest to lowest incidence are called `Pareto` charts.

```python
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.figure(figsize = (6,6))
plt.bar(company,profit)
plt.title("Vertical (column) bar chart presenting various companies and their profits on the y and x  axes respectively\n")
plt.xlabel("Profits")
plt.ylabel("Companies")
plt.show()
```

```python
plt.figure(figsize = (6,6))
plt.barh(company,profit, height = 0.6, )
plt.title("Horizontal bar chart presenting various companies and their profits on the y and x  axes respectively\n")
plt.xlabel("Profits")
plt.ylabel("Companies")
plt.show()
```

#### Q5: What is a box plot? Why is it used? Using the following data plot a box plot.

`box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200) `

#### Answer:

A box plot is a standardized way of visually representing the dataset based on the five-number summary: the minimum, the maximum, the sample median, and the first and third quartiles. 

1. Minimum (Q0 or 0th Percentile): The lowest data point in the dataset excluding any outliers.
2. Maximum (Q4 or 100th Percentile): The highest data point in the dataset excluding any outliers.
3. Median (Q2 or 50th Percentile): The middle value in the dataset. 
4. First Quartile (Q1 or 25th Percentile): also known as the lower quartile qn(0.25), it is the median of the lower half of the dataset. 
5. Third Quartile (Q3 or 75th Percentile): also known as the upper quartile qn(0.75), it is the median of the upper half of the dataset. 

Another important element is Interquartile range (IQR): the distance between the upper and lower quartiles. 

`IQR` = Q3 - Q1 = qn(0.75) - qn(0.25)

A box plot includes two parts: a box and a set of whiskers. The box is drawn from Q1 to Q3 with a horizontal line drawn in the middle to denote the median. The whiskers must end at an observer data point, but can be defined in various ways:

1. Boundary of lower whisker = lowest value of the dataset and Boundary of upper whisker = maximum value of the dataset. 
2. Boundary of lower whisker = 1.5*IQR below the lower the lower quartile(Q1) and Boundary of lower whisker = 1.5*IQR above the upper quartile (Q3). All other observed points outside the boundary of the whiskers are called `outliers`. The outliers can be plotted on the box-plot as a dot, a small circle, a star, etc.,
3. One standard deviation above and below the mean of the dataset. 
4. The 9th Percentile and the 91st percentile of the dataset. 
5. The 2nd percentile and the and the 98th Percentile  of the dataset. 

Note: The unusual percentiles 2%, 9%, 91%, 98% are sometimes used for whisker cross-hatches and whisker ends to depict the `seven-number summary`

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
# print(box1)
```

```python
fig, ax = plt.subplots(1,2)
fig.subplots_adjust(hspace = 0.8, top = 0.85, wspace = 0.6)
# Box Plot for box1 NumPy array
ax[0].set_title("Box Plot for box1 NumPy array")
ax[0].boxplot(box1)

# # Box Plot for box1 NumPy array
ax[1].set_title("Box Plot for box2 NumPy array")
ax[1].boxplot(box2)
plt.show()
```
