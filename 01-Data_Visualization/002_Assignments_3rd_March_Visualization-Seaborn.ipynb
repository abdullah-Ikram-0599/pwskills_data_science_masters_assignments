#### Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

5 plots that we can plot using the Seaborn library are:

##### 1. Scatter Plot (sns.scatterplot()):

A scatter plot is a graphical representation used to explain the relationship between two continuous variables in a dataset.

It uses a Cartesian coordinate system, where each data point is symbolized by a marker on a two-dimensional plane.

It represents the independent variable on the x-axis, and the dependent variable on the y-axis.
        
It can also be used to create complex plots such as scatter plot with additional features like colour or size to 
represent additional information.

By plotting multiple data points, it enables analysts and reserachers to observes the dispersion and clustering of data,         identifying potential relationships such as positive or negative correlations, clusters, or outliers.
    
Some of the key arguments of `sns.scatterplot()` are:

i. `data` provides the data to plot via a Pandas DataFrame.

ii. `x=` and `y=` provide the variables to plot on the x- and y-axis respectively.

iii. `hue=` adds an additional (categorical or numerical) variable to plot via a color mapping. 

iv. `style=` adds an additional (categorical) variable to plot via different styes (shapes).

##### 2. Relational Plot (sns.relplot()): --> Figure level function

A relational plot enables us to visualize the relationship between two quantitative (numeric) variables using either scatter plots or line plots. This can be customized using kind (kind = 'line', or kind = 'scatter'). 
        
All the above mentioned arguments (in sns.scatterplot()) except the following are also present in sns.relplot():

i. `kind=` determines what type of chart to create. `kind = 'scatter'` is the default value. Other value is `kind=line`

ii. `col=` and and iii.`row=`   will take the name of a DataFrame column (Series), and then create a new column or row (respectively) containing subplots for each unique value in that series. 

##### 3. Categorical Plot (sns.catplot()):

It is used to create different types of plots to help us analyze and understand the relationships between continuous and categorical variables. 

Some of its key arguments are: data, x, y, hue, row, col, kind. 

`row, col` arguments are categorical variables that will determine the facet of the grid.


It is a returns a figure-level interface for drawing categorical plots onto a FacetGrid. 

Its `kind=` parameter is used to determine which type of categorical plot to create. Its unique values are:

i. `Categorical Scatter plots:` stripplot() (with `kind="strip"`, the default) and swarmplot() (with `kind = "swarm"`).

ii. `Categorical Distribution plots`: boxplot (with `kind = "box"`), violinplot (with kind = `"violin"`), and boxenplot (with `kind = "boxen"`)

iii. `Categorical Estimate Plots`: pointplot (with `kind = "point"`), barplot (with `kind = "bar"`), and countplot (with `kind = "count"`)

##### 4. Joint Plot (sns.jointplot()):

It is used to draw and plot of two variables with bivariate and multivariate graphs. 

It creates a scatterplot with two histograms at the top and right margins of the graph by default. 

It is very helpful to have both univariate and bivariate plots in one figure. This is because the univariate analysis focuses on one variable, it describes, summarizes, and shows any patterns in our data and the bivariate analysis explores the relationship between two variables and also describes the strength of their relationship. 

Some of its key arguments are: data, x, y, hue, and kind.

Its `kind =` parameter is used to specify kind of plot to draw. Its unique values are: `"scatter" (default) | "kde" | "hist" | "hex" | "reg" | "resid" ` 

##### 5. Pair Plot (sns.pairplot()):

It is used to plot pairwise relationships in a dataset. 

By default, this function will create a grid of Axes such that each numeric variable in `data` will be shared across the y-axes across a single row and the x-axes across a single column. The diagonal plots are treated differently: a univariate distribution plot is drawn to show the marginal distribution of the data in each column. 

The simplest invocation uses `scatterplot()` for each pairing of the variables and `histplot()` for the marginal plots on the diagonal. 

Some of its key arguments are: data, hue, and kind, and diag_kind. 

Its `kind =` parameter is used to specify kind of plot to draw. It determines both the diagonal and off-diagonal plotting style. Its unique values are: `"scatter" | "kde" | "hist" | "reg"`

Its `diag_kind` parameter is used to determine the kind of plot for the diagonal subplots. Its unique values are `"auto" | "hist" | "kde" | "none". 

#### Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x = "timepoint" and y = "signal" for different events and regions.
#### Note: timepoint, signal, event, and region are columns in the fmri dataset.

```python
import seaborn as sns
```

```python
fmri = sns.load_dataset("fmri")
```

```python
fmri
```

```python
sns.relplot(data = fmri, x = "timepoint", y = "signal", hue = 'event', style = 'region', row ='event', col = "region", 
            kind = "line")
```

#### Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.
#### Note: pclass, age, and fare are columns in the titanic dataset.

```python
titanic = sns.load_dataset('titanic')
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(2, figsize = (8,6))

fig.set_size_inches(8,10)
fig.subplots_adjust(hspace = 0.4, top = 0.9)

fig.suptitle("Box Plots", fontsize = 15, ha = 'center')

sns.boxplot(data = titanic, x = 'pclass', y = 'age', ax = ax[0])
ax[0].title.set_text("Box plot pclass versus age\n")

sns.boxplot(data = titanic, x = 'pclass', y = 'fare',ax = ax[1])
ax[1].title.set_text("Box plot of pclass versus fare\n")
```
